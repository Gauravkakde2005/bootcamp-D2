```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.countplot(x='area', data=data)
plt.title('Distribution of Areas')
plt.xlabel('Area')
plt.ylabel('Count')
plt.show()
```

```python


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=20)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python

import matplotlib.pyplot as plt
least_expensive_data = data.sort_values('price').head(10)
plt.figure(figsize=(10, 6))
bars = plt.bar(least_expensive_data.index, least_expensive_data['price'], color='blue')


for i in range(5):
    bars[i].set_color('green')

plt.title('Top 10 Least Expensive Prices')
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='airconditioning', data=data)
plt.title('Count of Properties with Air Conditioning')
plt.xlabel('Air Conditioning')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(10, 5))
plt.scatter(data['area'], data['price'], alpha=0.6, color='blue', edgecolors='w', s=100 , cmap='viridis')
plt.title('Scatter Plot of Price vs Area')
plt.xlabel('Area (square units)')
plt.ylabel('Price')
plt.grid()
plt.show()
```
